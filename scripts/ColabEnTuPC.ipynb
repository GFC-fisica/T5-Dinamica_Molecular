## Paso 2

```python
!pip install colab_ssh --upgrade --quiet  
from colab_ssh import launch_ssh_cloudflared, init_git_cloudflared
```

## Paso 3

Esta celda inicializa un repositorio de Git y lo conecta al repositorio de Github. Para ello, es necesario proporcionar tu nombre de usuario y correo de Github, así como un token de acceso personal.

```python
github_repo = "https://github.com/GFC-fisica/T5-Dinamica_Molecular"
github_username = "" # Tu usuario de Github
github_email_address = "" # Tu correo de Github
github_access_token = "" # Tu token de acceso. Para crearlo, ve a "https://docs.github.com/en/authentication/keeping-your-account-and-data-secure/creating-a-personal-access-token#creating-a-personal-access-token-classic"
ssh_password = "" # Tu contraseña, puedes definirla como desees
  
launch_ssh_cloudflared(ssh_password)  
init_git_cloudflared(repository_url=github_repo + ".git",  
    personal_token=github_access_token,  
    branch="master",
    email=github_email_address,  
    username=github_username)
```
